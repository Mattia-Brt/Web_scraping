```python
import bs4, requests, webbrowser
import requests
import json
import pandas as pd
```

## Lista compagnie aeree europee

```python
url='https://airmundo.com/it/blog/elenco-delle-compagnie-aeree-in-europa/'
s = requests.session()
response = s.get(url)
response.raise_for_status()
soup=bs4.BeautifulSoup(response.text, 'html.parser')
lieu=[]
for td in soup.select('td'):
    a = td.find('a')
    if a is not None:
        #print (a.attrs['href'])
        print(a.contents[0])
        lieu.append(a.contents[0])
```

## Lista compagnie mondiali con codici iata

```python
url='https://www.flightradar24.com/data/airlines/'
headers = {'User-Agent': 'Safari'}

s = requests.session()
response = s.get(url, headers = headers)
response
```

```python
response.raise_for_status()
soup=bs4.BeautifulSoup(response.text, 'html.parser')
li=[]
for td in soup.select('td'):
    if len(td.text)>1 :
        if td.text != "":
            #print(td.text)
            li.append(td.text)
name = []
iata1 = []
aircraft = []
counter = 1

for i in li:
    if counter == 1: # siamo alla prima riga
        counter +=1
    elif counter == 2: # prendo il nome
        name.append(i)
        counter +=1
    elif counter == 3: # prendo il codice
        iata1.append(i)
        counter +=1
    else: # prendo il numero di aerei per la compagnia
        aircraft.append(i)
        counter = 1

iata = []
for i in iata1:
    s = str(i)
    s = s.replace(" / ", "-")
    s = s.lower()
    iata.append(s)
```

### Confronto lieu e se ci sono nella name ne prendo il codice iata e il numero di aircraft

```python
nameEu = []
iataEu = []
for i in name:
    if i in lieu: # se il nome è contenuto negli europei allora lo prendo
        indice = name.index(i)
        nameEu.append(i) #salvo il nome
        iataEu.append(iata[indice]) #salvo il codice iata corrispondente
```

### Creo il dataframe con tutte le info 

```python
nameFrame = pd.DataFrame(nameEu, columns =['Airline'])
iataFrame = pd.DataFrame(iataEu, columns =['iataAirline'])

df = pd.concat([nameFrame,iataFrame],axis=1)
df
```
